```python
# 중간에서부터 함수로 변환하여 결측치 대체 수행
# d2 : 일최고기온
# d3 : 일평균기온
# d5 : 일평균상대습도
# d6 : 일평균풍속 
# d7 : 일최대풍속

d2 = '일최고기온(°C)'
d3 = '일평균기온(°C)'
d5 = '일평균상대습도(%)'
d6 = '일평균풍속(m/s)'
d7 = '일최대풍속(m/s)'
 
# reg는 지역, for_day는 앞날짜, d는 위의 값, index는 결측치행번호
def replace_d(reg,for_day, d,index):
    replace = weather[(weather['지점'] == reg)&(weather['일시'] == for_day)][d]
    if d == d2:
        weather.iloc[index,2] = replace
    elif d == d3:
        weather.iloc[index,3] = replace
    elif d == d5:
        weather.iloc[index,5] = replace
    elif d == d6:
        weather.iloc[index,6] = replace
    elif d == d7:
        weather.iloc[index,7] = replace
```
